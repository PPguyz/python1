## Hi, This is a simple trial! We call the necessary libraries as the first step to our code:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
```

## We make a random variable (a):

```python
a = np.random.rand(100)*10

noise = np.random.normal(0,1,100)

# 0 is the mean of the normal distribution you are choosing from
# 1 is the standard deviation of the normal distribution
# 100 is the number of elements you get in array noise

b = 2*a + noise
```

## Sketching the graph results in:

```python
plt.plot(a,b, '.')
plt.xlabel("a")
plt.ylabel("b")
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = sp.stats.linregress(a, b)
```

```python
line = slope*a+intercept
r_value_cut = int(r_value/0.01)/100
slope_cut = int(slope/0.01)/100
intercept_cut = int(intercept/0.01)/100

plt.plot(a,b, '.')
plt.plot(a,line)
plt.xlabel("a")
plt.ylabel("b")
plt.text(0, 20, '$R^2$ = %s'%r_value_cut, fontsize = 10)
plt.text(0, 19, 'b = ${} a$ + {}'.format(slope_cut,intercept_cut), fontsize = 10)


plt.show()
```

That's it for the first tutorial! We tried to make a random variable (a) and a dependent variable (b).
Using regression, the best line fit is obtained!
